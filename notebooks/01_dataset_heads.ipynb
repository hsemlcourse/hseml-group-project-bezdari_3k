# Первые 10 строк каждого CSV-датасета

В этом ноутбуке каждый датасет открыт в отдельной ячейке. Используется `pandas.read_csv(..., nrows=10)` и затем `.head(10)`, чтобы не загружать огромные CSV целиком в память.

`numpy` тоже импортирован, но метод `head()` относится к pandas, не к NumPy.

```python
from pathlib import Path

import numpy as np
import pandas as pd

pd.set_option("display.max_columns", 80)
pd.set_option("display.max_rows", 20)
pd.set_option("display.max_colwidth", 80)

PROJECT_ROOT = Path.cwd()
if not (PROJECT_ROOT / "data").exists():
    PROJECT_ROOT = PROJECT_ROOT.parent

DATA_DIR = PROJECT_ROOT / "data"
DATA_DIR
```

```python
def show_head(file_name: str, n_rows: int = 10) -> pd.DataFrame:
    path = DATA_DIR / file_name
    df = pd.read_csv(path, nrows=n_rows)
    print(f"{file_name}: первые {len(df)} строк, показано {df.shape[1]} колонок")
    return df.head(n_rows)
```

## `correlations.csv`

**Что это за датасет:** корреляционная матрица для полной Featuretools-матрицы.

**За что отвечает:** показывает, насколько признаки линейно связаны друг с другом. Значения лежат примерно от `-1` до `1`: ближе к `1` - сильная положительная связь, ближе к `-1` - сильная отрицательная связь, около `0` - слабая линейная связь.

**Как читать колонки:** названия колонок - это признаки из исходной feature matrix. Каждая строка тоже соответствует признаку, но в этом файле имя строки не вынесено отдельной колонкой, поэтому первые значения выглядят как корреляции признаков между собой.

**Для чего используем:** для EDA, поиска признаков, связанных с `TARGET`, и поиска сильно похожих признаков. Для обучения моделей этот файл не используем.

```python
show_head("correlations.csv")
```

## `correlations_spec.csv`

**Что это за датасет:** корреляционная матрица для компактного `spec`-набора признаков.

**За что отвечает:** делает то же самое, что `correlations.csv`, но для меньшего набора признаков. Это удобнее для анализа, потому что колонок меньше.

**Как читать колонки:** колонка `Unnamed: 0` содержит название признака-строки. Остальные колонки - признаки, с которыми считается корреляция. Например, значение на пересечении строки `EXT_SOURCE_3` и колонки `TARGET` показывает связь `EXT_SOURCE_3` с дефолтом.

**Для чего используем:** для EDA и объяснения зависимостей в compact/spec-наборе. Для обучения моделей напрямую не используем.

```python
show_head("correlations_spec.csv")
```

## `feature_importances.csv`

**Что это за датасет:** таблица важностей признаков для основной матрицы `feature_matrix.csv`.

**За что отвечает:** показывает, какие признаки оказались наиболее полезными в модели автора датасета. Чем выше `importance`, тем сильнее признак помогал модели.

**Как читать колонки:** `index` - исходный индекс признака, `feature` - название признака, `importance` - численная важность признака.

**Для чего используем:** для отбора top-N признаков, интерпретации и объяснения, какие признаки важны для кредитного риска. Сам по себе файл не содержит клиентов и `TARGET`, поэтому для обучения как X/y он не подходит.

```python
show_head("feature_importances.csv")
```

## `feature_matrix.csv`

**Что это за датасет:** основная Featuretools-признаковая матрица.

**За что отвечает:** хранит данные для машинного обучения: каждая строка - отдельный клиент/кредитная заявка, каждая колонка - признак клиента или агрегированный признак из связанных таблиц Home Credit.

**Как читать ключевые колонки:** `SK_ID_CURR` - идентификатор клиента/заявки, `TARGET` - целевая переменная, `0` означает нет дефолта, `1` означает дефолт, `-999` означает Kaggle test row без разметки. Остальные колонки - признаки: доход, сумма кредита, регион, возраст, телефонные флаги, агрегаты по прошлым кредитам и т.д.

**Для чего используем:** можно использовать для обучения моделей, но файл большой. В нашем CP1 вместо него выбран более лёгкий `feature_matrix_spec.csv`.

```python
show_head("feature_matrix.csv")
```

## `feature_matrix_advanced.csv`

**Что это за датасет:** расширенная advanced-версия признаковой матрицы.

**За что отвечает:** содержит больше признаков, чем обычная `feature_matrix.csv`: автоматические Featuretools-агрегаты, дополнительные условия, а также ручные признаки вроде кредитных ratios.

**Как читать ключевые колонки:** `SK_ID_CURR` - идентификатор заявки, `TARGET` - целевая переменная. В этом файле test rows часто имеют `TARGET = NaN`, а не `-999`. Колонки с `NEW_...` - новые engineered-признаки, например отношение аннуитета к доходу. Колонки вида `MEAN(...)`, `MAX(...)`, `SUM(...)` - агрегаты из связанных таблиц.

**Для чего используем:** подходит для более продвинутых экспериментов, но очень широкий и тяжёлый. Для CP1 можно упомянуть как альтернативный источник признаков, но основной пайплайн лучше держать на `feature_matrix_spec.csv`.

```python
show_head("feature_matrix_advanced.csv")
```

**За что отвечает:** показывает вариант подготовленных признаков, который использовался в tutorial/article-подходе. По смыслу это тоже обучающая таблица: строки - клиенты, колонки - признаки.

Финансовые признаки вроде `AMT_ANNUITY`, `AMT_CREDIT`, `AMT_GOODS_PRICE`, `AMT_INCOME_TOTAL` описывают размер кредита, платежа, цены товара и доход клиента. Колонки `AMT_REQ_CREDIT_BUREAU_*` показывают число запросов в кредитное бюро за разные периоды.

**Для чего используем:** можно использовать как альтернативную обучающую матрицу или для сравнения с другими версиями feature matrix. В нашем CP1 это вспомогательный файл, не основной.

## `feature_matrix_spec.csv`

**Что это за датасет:** компактная spec-признаковая матрица.

**За что отвечает:** это главный датасет для нашего CP1. Он содержит достаточно признаков для обучения, но при этом заметно легче, чем advanced/full версии.

**Как читать ключевые колонки:** `SK_ID_CURR` - идентификатор клиента/заявки, `TARGET` - целевая переменная. Значения `0` и `1` используются для обучения, строки с `TARGET = -999` - это Kaggle test set без известного ответа, их удаляем перед supervised learning. `ORGANIZATION_TYPE`, `NAME_INCOME_TYPE` и похожие колонки - категориальные признаки. `AMT_*`, `DAYS_*`, `EXT_SOURCE_*`, `FLAG_*` - числовые/бинарные признаки клиента и заявки.

**Для чего используем:** основной источник X/y для pipeline: очистка, split, feature engineering, baseline и все эксперименты моделей.

```python
show_head("feature_matrix_spec.csv")
```

**Как читать колонки:** `Unnamed: 0` - порядковый индекс строки, `feature` - название признака, `importance` - важность. Большие значения `importance` означают, что признак чаще/сильнее использовался моделью.

**Для чего используем:** для анализа advanced-признаков и возможного отбора top-N признаков, если решим строить модель на `feature_matrix_advanced.csv`. Для текущего CP1 это справочный файл.

## `spec_feature_importances_ohe.csv`

**Что это за датасет:** таблица важностей признаков для `feature_matrix_spec.csv` после one-hot encoding.

**За что отвечает:** помогает понять, какие признаки из compact/spec-набора наиболее полезны. Именно этот файл мы используем в коде, чтобы выбирать top-N признаков и не загружать все 885 колонок каждый раз.

**Как читать колонки:** `Unnamed: 0` - порядковый индекс строки, `feature` - название признака, `importance` - важность признака. Если `importance = 0`, признак почти не помогал модели в расчётах автора.

**Для чего используем:** для нашего CP1 это основной файл отбора признаков: функция `load_top_features()` читает его, сортирует признаки по `importance` и выбирает самые важные колонки из `feature_matrix_spec.csv`.

```python
show_head("spec_feature_importances_ohe.csv")
```
